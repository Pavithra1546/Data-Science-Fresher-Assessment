# Box Plot

```python
#boxplot-means percentile
sb.boxplot(x="smoker",y="total_bill",data=dataset,hue="sex")
plt.show()
```

```python
#boxplot-means percentile
sb.boxplot(x="day",y="total_bill",data=dataset,hue="sex")
plt.show()
```

```python
#boxplot-means percentile
sb.boxplot(x="time",y="total_bill",data=dataset,hue="sex")
plt.show()
```
